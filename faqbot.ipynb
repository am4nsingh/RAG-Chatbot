```python
from langchain.document_loaders import DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    path=r"C:\Users\amans\Desktop\Chatbot_project\UX4G",
    glob="**/*.pdf",
    loader_cls=PyPDFLoader
)

all_documents = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema.document import Document
splitter = RecursiveCharacterTextSplitter(separators=["\n\n"])
split_docs = splitter.split_documents(all_documents)
docs =[]
for d in split_docs:
    if "Capacity building.pdf" in str(d.metadata["source"]):
        docs.append(Document(page_content = d.page_content, metadata ={"type":"Capacity Building"}))
    elif "Design Handbook.pdf" in str(d.metadata["source"]):
        docs.append(Document(page_content = d.page_content, metadata ={"type":"Design Handbook"}))
    elif "Design system.pdf" in str(d.metadata["source"]):
        docs.append(Document(page_content = d.page_content, metadata ={"type":"Design system"}))
    elif "GENERAL.pdf" in str(d.metadata["source"]):
        docs.append(Document(page_content = d.page_content, metadata ={"type":"GENERAL"}))
    elif "U4XG.pdf" in str(d.metadata["source"]):
        docs.append(Document(page_content = d.page_content, metadata ={"type":"UX4G"}))
for d in docs:
    print(d,"\n***************************************************************************************\n")
```

```python
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
persist_directory = r"C:\Users\amans\Desktop\Chatbot_project\faq_vdb"
embeddings = HuggingFaceEmbeddings(model_name = "BAAI/bge-base-en-v1.5")
vectordb = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory=persist_directory
)
```
